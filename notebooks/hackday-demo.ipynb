```python
from mast_aladin.hackday import MastManager
import warnings
```

### Create a MastManager
The `MastManager` (name TBD) is responsible for managing the state of the UI, applications, and various plugins (name TBD) that support interoperable features between apps. 

For this demo, we are creating only a simple template which loads a `MastAladin` and `Imviz` app instances and places them in separate Sidecars. A `Viewer Sync` Plugin it also created, for now we are placing it inline with the notebook for simplicity

```python
mm = MastManager('default_template.json')
```

### Load data into the applications
We can now access the applications created through the `MastManager.apps` property. We can then load data and initial parameters into those applications as usual.

Note: don't forget to align Imviz by WCS after loading data in as the plugin won't work without it

```python
aladin = mm.apps["mast-aladin"]
aladin.target = "Cartwheel Galaxy"
aladin.fov = .1

imviz = mm.apps["imviz"]
filenames = [
    'jw02727-o002_t062_nircam_clear-f277w_i2d.fits',
]

with warnings.catch_warnings(), imviz.batch_load():
    warnings.simplefilter('ignore')
    for fn in filenames:
        uri = f"mast:JWST/product/{fn}"
        imviz.load(uri, cache=True)
imviz.link_data(align_by='wcs')
```
